```python
#Normalization
from sklearn import preprocessing
min_max_scaler=preprocessing.MinMaxScaler()
# Create an object to transform the data to fit minmax processor
x_scaled=min_max_scaler.fit_transform(dataset1[['Home Team Goals']].values.astype(float))
#Run the normalizer on the dataframe
df_normalized=pd.DataFrame(x_scaled)
dataset1['normalized2']=x_scaled
dataset1.sort_values(by='Home Team Goals')
dataset1[['Home Team Goals','normalized2']]
```

```python
#Z-score normalisation

from scipy import stats
dataset1['zscore']=stats.zscore(dataset1['Home Team Goals'])
dataset1.sort_values(by='zscore')
dataset1[['zscore','Home Team Goals']]
```
